```python
import matplotlib.pyplot as plt
import random
vertices = [(0, 0), (1, 0), (0.5, 0.866)]
def isInside(x1, y1, x2, y2, x3, y3, x, y):
    def area(x1, y1, x2, y2, x3, y3):
        return abs((x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)) / 2.0)
    A = area(x1, y1, x2, y2, x3, y3)
    A1 = area(x, y, x2, y2, x3, y3)
    A2 = area(x1, y1, x, y, x3, y3)
    A3 = area(x1, y1, x2, y2, x, y)
    return A == A1 + A2 + A3
seed_x = 0.25
seed_y = 0.25
num_steps = 10000
if not isInside(0, 0, 1, 0, 0.5, 0.866, seed_x, seed_y):
    raise ValueError("N/A")
seed = (seed_x, seed_y)
points = [seed]
for i in range(num_steps):
    next_vertex = vertices[random.randint(0, 2)]
    last_point = points[-1]
    next_point = ((last_point[0] + next_vertex[0]) / 2, (last_point[1] + next_vertex[1]) / 2)
    points.append(next_point)
def plot_solution(points):
    plt.scatter([p[0] for p in points[12:]], [p[1] for p in points[12:]], s=0.1)
    plt.title("Sierpinski Triangle")
    plt.show()
plot_solution(points)
```
