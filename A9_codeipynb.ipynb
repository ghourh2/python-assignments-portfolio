```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
#  4(a)
x, y = sp.symbols('x y', real=True)
f = sp.exp(x)*sp.sin(y) + y**3
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)
print("f(x,y)       =", f)
print("∂f/∂x        =", df_dx)
print("∂f/∂y        =", df_dy)
print("------")
#  4(b)
g = x**2 * y + x * y**2
dg_dx = sp.diff(g, x)
dg_dy = sp.diff(g, y)
grad_g_at_1_neg1 = (dg_dx.subs({x:1, y:-1}), dg_dy.subs({x:1, y:-1}))
magnitude_grad_g = sp.sqrt(grad_g_at_1_neg1[0]**2 + grad_g_at_1_neg1[1]**2)
print("g(x,y)             =", g)
print("∂g/∂x              =", dg_dx)
print("∂g/∂y              =", dg_dy)
print("∇g(1, -1)          =", grad_g_at_1_neg1)
print("‖∇g(1, -1)‖        =", magnitude_grad_g)
print("------")
#  4(c)
h = sp.log(x**2 + y**2)
dh_dx = sp.diff(h, x)
dh_dy = sp.diff(h, y)
dh_dxdx = sp.diff(h, (x, 2))
dh_dydy = sp.diff(h, (y, 2))
dh_dxdy = sp.diff(sp.diff(h, x), y)
dh_dydx = sp.diff(sp.diff(h, y), x)
print("h(x,y)              =", h)
print("∂h/∂x               =", dh_dx)
print("∂h/∂y               =", dh_dy)
print("∂²h/∂x²             =", dh_dxdx)
print("∂²h/∂y²             =", dh_dydy)
print("∂²h/∂x∂y            =", dh_dxdy)
print("∂²h/∂y∂x            =", dh_dydx)
print("Mixed partials match:", dh_dxdy.equals(dh_dydx))
print("------")
#  4(d)
j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z_j = j_func(X, Y)
plt.figure()
plt.contourf(X, Y, Z_j, levels=50, cmap='viridis')
plt.colorbar()
plt.title("Contour plot of j(x, y) = x^3 - 3xy + y^3")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
print("------")
#  4(e)
k = sp.sin(sp.sqrt(x**2 + y**2))
k_func = sp.lambdify((x, y), k, 'numpy')
Z_k = k_func(X, Y)
plt.figure()
plt.contourf(X, Y, Z_k, levels=50, cmap='plasma')
plt.colorbar()
plt.title("Contour plot of k(x, y) = sin( sqrt(x^2 + y^2) )")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
