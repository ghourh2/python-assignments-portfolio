```python
import numpy as np
import matplotlib.pyplot as plt

# Q3
def normal_density(mean, variance, x):
    sigma = np.sqrt(variance)
    coefficient = 1.0 / (np.sqrt(2 * np.pi) * sigma)
    exponent = - ((x - mean) ** 2) / (2 * variance)
    return coefficient * np.exp(exponent)
def integrate_normal_density(mean, variance, a, b, num_points=10000):
    x_values = np.linspace(a, b, num_points)
    y_values = normal_density(mean, variance, x_values)
    area = np.trapezoid(y_values, x_values)
    return area
x_std = np.linspace(-4, 4, 1000)
x_male = np.linspace(140, 210, 1000)  
#standard normal:mean=0,variance=1
#male heights:mean=171,variance=(7.1^2)
y_std = normal_density(0, 1, x_std)
male_mean = 171
male_variance = 7.1 ** 2
y_male = normal_density(male_mean, male_variance, x_male)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x_std, y_std, label='N(0,1)')
plt.title('Density For Standard Normal')
plt.subplot(1, 2, 2)
plt.plot(x_male, y_male, color='red', label='N(171,7.1^2)')
plt.title('Density for Male Height')
plt.xlabel('Height')
plt.tight_layout()
plt.show()
probability = integrate_normal_density(male_mean, male_variance, 162, 190)
print("probability: {:.4f}".format(probability))
```
